```python
from askesis import load_db, viz
import matplotlib.pyplot as plt

# Configure matplotlib
%matplotlib inline
viz.setup_style()
```

## Load Your Data

Export your data from Settings in the Askesis app, then load it here:

```python
# Update this path to your exported .db file
db = load_db("~/Downloads/askesis-2024-03-09.db")
db
```

## Dashboard Overview

```python
fig = viz.dashboard(db)
plt.show()
```

## Weight Analysis

```python
# Weight trend
viz.plot_weight_trend(db)
plt.show()
```

```python
# Weight statistics
weight = db.weight_series()
print(f"Weight range: {weight.min():.1f} - {weight.max():.1f} kg")
print(f"Average: {weight.mean():.1f} kg")
print(f"Change: {weight.iloc[-1] - weight.iloc[0]:.1f} kg")
```

## Sleep Analysis

```python
viz.plot_sleep_trend(db)
plt.show()
```

```python
# Sleep statistics
sleep = db.sleep_series()
print(f"Average sleep: {sleep.mean():.1f} hours")
print(f"Days under 7 hours: {(sleep < 7).sum()}")
print(f"Days over 8 hours: {(sleep >= 8).sum()}")
```

## Activity Analysis

```python
# Activity summary by type
db.activity_summary()
```

```python
# Activities by day of week
activities = db.activities.copy()
activities['day_of_week'] = activities['date'].dt.day_name()
activities.groupby('day_of_week').size().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).plot(kind='bar', title='Activities by Day of Week')
plt.xticks(rotation=45)
plt.show()
```

## Body Measurements Progress

```python
viz.plot_measurement_progress(db)
plt.show()
```

```python
# Measurement changes
db.measurement_changes()
```

## Custom Analysis

Use pandas for custom queries:

```python
# Example: Find best sleep days
db.daily_logs.nlargest(5, 'sleep_hours')[['date', 'sleep_hours', 'feelings']]
```

```python
# Example: Correlation between sleep and steps
logs = db.daily_logs.dropna(subset=['sleep_hours', 'steps'])
if len(logs) > 5:
    correlation = logs['sleep_hours'].corr(logs['steps'])
    print(f"Correlation between sleep and steps: {correlation:.2f}")
```
